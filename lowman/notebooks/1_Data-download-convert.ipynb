```python
#read the wkt file
lowman_wkt = gpd.read_file('./lowman/asf_wkt.geojson').to_wkt().loc[0, 'geometry']
print(lowman_wkt)
```

```python
# define search parameters for sierra flight line
start_date = datetime.strptime('2021-01-01', '%Y-%m-%d') 
end_date = datetime.strptime('2021-03-30', '%Y-%m-%d') 
```

```python
# search for data
results = asf.search(platform = 'UAVSAR',
                     processingLevel = (['INTERFEROMETRY_GRD']),
                     intersectsWith = lowman_wkt,
                     start = start_date,
                     end = end_date)

print(f'{len(results)} product(s) found')
```
