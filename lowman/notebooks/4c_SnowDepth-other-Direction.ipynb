```python
#Read the incidence angle tiff file
incidence_angle = rioxarray.open_rasterio('/SNOWDATA/Nah/lowman/Incidence_Angle/lowman_05208_21012_004_210210_L090_CX_01.inc.tiff')
```

```python
snowdepth_feb10_Mar03 = (-1 * unw_arr_whisker * 24.0)/(4 * np.pi * (np.cos(incidence_angle)  - np.sqrt(1.5 - (np.sin(incidence_angle)**2))))
```
