In this notebook, I plot lidar differenced DEMS snow depth. The Difference DEM data was a subtraction of snow-on and snow-off lidar data.

```python
import os
import numpy as np
import rioxarray
import matplotlib.pyplot as plt
```

```python
#Read the lidar data
lid_arr = rioxarray.open_rasterio('/home/naheemadebisi/snow-analytics/Western-Snow-Conference/QSI_DEMs/MCS_15Mar21_snowdepth_diff.tif')
```

```python
#select valid values
Lidar_snow = lid_arr.where(((lid_arr < float(lid_arr.quantile(0.95).values)) & (lid_arr > float(lid_arr.quantile(0.05).values))), other = np.nan)
```

```python
# Set font size and font family
plt.rcParams.update({'font.size': 18})
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

#create a figure and axes elements
fig, ax = plt.subplots(figsize=(12,10), constrained_layout=True)

#plot lidar differenced DEMs
Lidar_snow.plot(ax=ax, cmap = 'viridis' )

#set title
ax.set_title('')


plt.show()
```
