In this notebook, I plot multitemporal snow depth estimated from UAVSAR data. 

```python
#Files/directory package
import os

#plotting packages
import matplotlib.pyplot as plt
import seaborn as sns

#spatial data and other packages
import pandas as pd
import geopandas as gpd
import rioxarray 
#import earthpy as et
import numpy as np
#import rasterio
#from rasterio.warp import calculate_default_transform, reproject, Resampling
import glob
```

```python
#read the raster files opf snow depth
lowman_15_20Jan = rioxarray.open_rasterio('/home/naheemadebisi/snow-analytics/lowman/MCS2/MCS_15_20Jan2021_snow_diff.tif')
lowman_20_27Jan = rioxarray.open_rasterio('/home/naheemadebisi/snow-analytics/lowman/MCS2/MCS_20_27Jan2021_snow_diff.tif')
lowman_27_03feb = rioxarray.open_rasterio('/home/naheemadebisi/snow-analytics/lowman/MCS2/MCS_27_03Feb2021_snow_diff.tif')
lowman_03_10feb = rioxarray.open_rasterio('/home/naheemadebisi/snow-analytics/lowman/MCS2/MCS_03_10Feb2021_snow_diff.tif')
lowman_10_03mar = rioxarray.open_rasterio('/home/naheemadebisi/snow-analytics/lowman/MCS2/MCS_feb10_03Mar2021other_snow_diff.tif')
lowman_03_10mar = rioxarray.open_rasterio('/home/naheemadebisi/snow-analytics/lowman/MCS2/MCS_03_10Mar2021_snow_diff.tif')
lowman_10_16mar = rioxarray.open_rasterio('/home/naheemadebisi/snow-analytics/lowman/MCS2/MCS_10_16Mar2021_snow_diff.tif')
lowman_16_22mar = rioxarray.open_rasterio('/home/naheemadebisi/snow-analytics/lowman/MCS2/MCS_16_22Mar2021_snow_diff.tif')
```

```python
#Put the raster files into a list
arr_list_clip_reproj = [lowman_15_20Jan, lowman_20_27Jan, lowman_27_03feb, lowman_03_10feb, lowman_10_03mar, lowman_03_10mar, lowman_10_16mar, lowman_16_22mar]
```

```python
# Set font size and font family
plt.rcParams.update({'font.size': 18})
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

```

```python
#create a figure and axes elements
fig, ax = plt.subplots(figsize=(16,12), constrained_layout=True)

arr_list_clip_reproj[0].plot(ax=ax, cmap='Blues', robust = True, cbar_kwargs={'label': 'Snow depth (m)'})
ax.set_title('Lowman MCS15-20-21')

plt.show()
```

```python
#create a figure and axes elements
fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(16,12), constrained_layout=True)

arr_list_clip_reproj[0].plot(ax=ax[0,0], cmap='Blues', robust = True, cbar_kwargs={'label': 'Snow depth change (cm)'})
ax[0,0].set_title('Jan 15 - Jan 20')
arr_list_clip_reproj[1].plot(ax=ax[0,1], cmap='Blues', robust = True, cbar_kwargs={'label': 'Snow depth change (cm)'})
ax[0,1].set_title('Jan 20 - Jan 27')
arr_list_clip_reproj[2].plot(ax=ax[0,2], cmap='Blues', robust = True, cbar_kwargs={'label': 'Snow depth change(cm)'})
ax[0,2].set_title('Jan 27 - Feb 03')
arr_list_clip_reproj[3].plot(ax=ax[0,3], cmap='Blues', robust = True, cbar_kwargs={'label': 'Snow depth change (cm)'})
ax[0,3].set_title('Feb 03 - Feb 10')
arr_list_clip_reproj[4].plot(ax=ax[1,0], cmap='Blues', robust = True, cbar_kwargs={'label': 'Snow depth change (cm)'})
ax[1,0].set_title('Feb 10 - Mar 03')
arr_list_clip_reproj[5].plot(ax=ax[1,1], cmap='Blues', robust = True, cbar_kwargs={'label': 'Snow depth change (cm)'})
ax[1,1].set_title('Mar 03 - Mar 10')
arr_list_clip_reproj[6].plot(ax=ax[1,2], cmap='Blues', robust = True, cbar_kwargs={'label': 'Snow depth change (cm)'})
ax[1,2].set_title('Mar 10 - Mar 16')
arr_list_clip_reproj[7].plot(ax=ax[1,3], cmap='Blues', robust = True, cbar_kwargs={'label': 'Snow depth change (cm)'})
ax[1,3].set_title('Mar 16 - Mar 22')

#save the figure
plt.savefig('/home/naheemadebisi/snow-analytics/lowman/MCS2/MCS_snowdepth_timeseries.jpeg', dpi=500)
plt.show()

```
