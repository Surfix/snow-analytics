In this notebook, I modified the mosaic_function in 2 to allow searching for files with any extension format.

```python
#import packages
import numpy as np
import rasterio
from rasterio.merge import merge
from rasterio.plot import show
import glob
import os
```

```python
#function to merge rasters
def merge_rasters(search_dir, search_pattern, out_fp):
    """
    Tile a bunch of raster files
    """
    
    ##change to working directory
    #os.chdir(search_dir)

    #recursively searh for all files in working directory
    files = glob.glob(search_dir + search_pattern)
    
    #create empty list to store raster files
    src_files_to_mosaic = []

    #loop through files and append to list
    for file in files:
        src = rasterio.open(file)
        src_files_to_mosaic.append(src)
    
    #merge the rasters
    mosaic_raster, out_trans_raster = merge(src_files_to_mosaic)

    #copy the metadata
    out_meta = src.meta.copy()

    #update the metadata
    out_meta.update({
        'driver': 'GTiff',
        'height': mosaic_raster.shape[1],
        'width': mosaic_raster.shape[2],
        'transform': out_trans_raster
    })

    #Write the mosaic raster to disk
    with rasterio.open(out_fp, 'w', **out_meta) as dest:
        dest.write(mosaic_raster)
```
