In this notebook, I perform summary plot of the histogram of the data

```python
import numpy as np
import matplotlib.pyplot as plt
import rioxarray
import seaborn as sns
```

```python
#read the raster data
VH = rioxarray.open_rasterio('/home/naheemadebisi/snow-analytics/lidar/results/Mores/Mores_VH.tif', masked = True)
```

```python
# Set font size and font family of matplotlib for plotting
plt.rcParams.update({'font.size': 18})
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

#create a figure and axes elements
fig, ax = plt.subplots(figsize=(10,8), constrained_layout=True)

#plot the histogram of the raster data
plt.hist(VH.values.flatten(), bins = 100, range= (1,40))
plt.xlabel('Vegetation Height (m)')
plt.ylabel('Frequency')
plt.title('Histogram of Vegetation Height (m)')
#set the x-axis limits
#plt.xlim(1, 30)
#plt.grid(True)
plt.show()

#save the plot
fig.savefig('/home/naheemadebisi/snow-analytics/lidar/results/Mores/Mores_histogram.jpeg', dpi=500)
```
